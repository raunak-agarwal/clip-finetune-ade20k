### Scene Categories from ADE20K

```python
import os
import sys
import re

scenes_train_dir = '../dataset-exploration/clip-finetune-ade20k/data/ADE20K-pairs/imgs/training'
scenes_validation_dir = '../dataset-exploration/clip-finetune-ade20k/data/ADE20K-pairs/imgs/validation'
```

```python
def get_scene_distribution(scenes_dir):
    scenes = dict()
    for scene_folder in os.listdir(scenes_dir):
        scenes[scene_folder] = set()
        for subscene_folder in os.listdir(scenes_dir + '/' + scene_folder):
            for file_name in os.listdir(scenes_dir + '/' + scene_folder + '/' + subscene_folder):
                file_id = re.split('\_|\.', file_name)[2]
                scenes[scene_folder].add(file_id)
        scenes[scene_folder] = len(scenes[scene_folder])
    return scenes
            
        
```

```python
scenes_train = get_scene_distribution(scenes_train_dir)
scenes_validation = get_scene_distribution(scenes_validation_dir)
```

```python
import matplotlib.pyplot as plt

def draw_scene_distribution(scenes):
    scene_names = scenes.keys()
    scene_counts = scenes.values()
    x_pos = [i for i, _ in enumerate(scene_names)]

    plt.barh(x_pos, scene_counts, color='green')
    plt.ylabel("Scene name")
    plt.xlabel("Number of images")
    plt.title("Distribution of scenes from ADE20K")

    plt.yticks(x_pos, scene_names)

    plt.show()
```

```python
draw_scene_distribution(scenes_train)
```

```python
draw_scene_distribution(scenes_validation)
```
